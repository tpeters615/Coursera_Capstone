```python
import pandas as pd 
```

```python
df_PNB = pd.read_csv('Part1_PBN.csv')
df_PNB
```

```python
geo_data = pd.read_csv('https://cocl.us/Geospatial_data')
#geo_data=geo_data[geo_data['Postal Code'].isin(['M5G', 'M2H', 'M4B', 'M1J', 'M4G', 'M4M', 'M1R', 'M9V', 'M9L', 'M5V', 'M1B', 'M5A'])]
geo_data
```

```python
PNB=df_PNB.sort_values('PostalCode',ascending=True)
geo=geo_data.sort_values('Postal Code', ascending=True)
```

```python
final=pd.merge(PNB, geo, left_on='PostalCode', right_on='Postal Code')
final=final.drop(['Unnamed: 0', 'Postal Code'], axis=1)
```

```python
final.to_csv('full_df.csv')
```
